```python
list_dict_locations = []
for url in unquote_locations_url:
    longitude , latitude= re.findall(r'!3d(.*)\!4d(.*)\?', url)[0]
    name = re.findall(r'place\/(.*)\/data', url)[0]
    list_dict_locations.append({"name": name.replace("+", "_").replace("\u202d",""),
                                "longitude" : longitude,
                                "latitude" : latitude,
                                "url" : url})
```
